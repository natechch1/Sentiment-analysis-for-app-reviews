```python
# Task 2
import pprint
!pip install google-play-scraper
!pip install nltk
!mkdir Task2

import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
import re
import inflect

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(review):
  # Remove punctuation
  review = review.translate(str.maketrans('', '', string.punctuation))
  review = review.translate(str.maketrans('', '', string.punctuation))
  # Remove special characters and emojis
  pattern = re.compile("["
                          u"\U0001F600-\U0001F64F" # emoticons
                          u"\U0001F300-\U0001F5FF" # symbols & pictographs
                          u"\U0001F680-\U0001F6FF" # transport & map symbols
                          u"\U0001F1E0-\U0001F1FF" # flags (iOS)
                          u"\U00002702-\U000027B0"
                          u"\U000024C2-\U0001F251"
                          "]+", flags=re.UNICODE)
  review = pattern.sub('', review)
  review = re.sub(r'[^\w\s]', '', review)

  # turn numbers into text
  p = inflect.engine()
  words = re.findall(r'\d+|\D+', review)
  review = ' '.join([p.number_to_words(word) if word.isdigit() else word for word in words])
 
  # remove extra white spaces
  review = ' '.join(review.split())
  
  # turn all words into lowercase
  review = review.lower()
  
  # remove stop words
  stop_words = set(stopwords.words('english'))
  tokens = word_tokenize(review)
  filtered_tokens = [word for word in tokens if word not in stop_words]
  review = ' '.join(filtered_tokens)

  # lemmatize the text
  lemmatizer = WordNetLemmatizer()
  tokens = word_tokenize(review)
  lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
  review = ' '.join(lemmatized_tokens)

  return review

## My SUD app Birday
# Read in the CSV file containing the reviews
reviews_df = pd.read_csv('Task1/com.minar.birday.csv', encoding='utf-8')

# open a CSV file and store pre-processed reviews 
with open('Task2/preprocessed_reviews_birday.csv', mode='w', newline='', encoding='utf-8') as csv_file:
  fieldnames = ['review']
  writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
  writer.writeheader()
  # Count review number and for output 15 sample reviews
  count = 0
  print("My SUD app (Birday) 15 sample pre-processed review:\n")
  # write each review to a separate row in the CSV file
  for review in reviews_df['Review Text']:
    # Apply preprocessing function to reviews
    temp = preprocess_text(review)
    count +=1
    writer.writerow({'review': temp})
    if count <= 15:
      print(str(count)+": "+temp+"\n")
print("\n")

## My similar app1 KIC
# Read in the CSV file containing the reviews
reviews_df = pd.read_csv('Task1/com.kic.kicapp.csv', encoding='utf-8')

# open a CSV file and store pre-processed reviews 
with open('Task2/preprocessed_reviews_kic.csv', mode='w', newline='', encoding='utf-8') as csv_file:
  fieldnames = ['review']
  writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
  writer.writeheader()
  # Count review number and for output 15 sample reviews
  count = 0
  print("My similar app1 (KIC) 15 sample pre-processed review:\n")
  # write each review to a separate row in the CSV file
  for review in reviews_df['Review Text']:
    # Apply preprocessing function to reviews
    temp = preprocess_text(review)
    count +=1
    writer.writerow({'review': temp})
    if count <= 15:
      print(str(count)+": "+temp+"\n")
print("\n")

## My similar app2 Enjoyable Food Tracker (youate)
# Read in the CSV file containing the reviews
reviews_df = pd.read_csv('Task1/com.youate.android.csv', encoding='utf-8')

# open a CSV file and store pre-processed reviews 
with open('Task2/preprocessed_reviews_youate.csv', mode='w', newline='', encoding='utf-8') as csv_file:
  fieldnames = ['review']
  writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
  writer.writeheader()
  # Count review number and for output 15 sample reviews
  count = 0
  print("My similar app2 (youate) 15 sample pre-processed review:\n")
  # write each review to a separate row in the CSV file
  for review in reviews_df['Review Text']:
    # Apply preprocessing function to reviews
    temp = preprocess_text(review)
    count +=1
    writer.writerow({'review': temp})
    if count <= 15:
      print(str(count)+": "+temp+"\n")
print("\n")

## My competitor app1 Discord
# Read in the CSV file containing the reviews
reviews_df = pd.read_csv('Task1/com.discord.csv', encoding='utf-8')

# open a CSV file and store pre-processed reviews 
with open('Task2/preprocessed_reviews_discord.csv', mode='w', newline='', encoding='utf-8') as csv_file:
  fieldnames = ['review']
  writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
  writer.writeheader()
  # Count review number and for output 15 sample reviews
  count = 0
  print("My competitor app1 (Discord) 15 sample pre-processed review:\n")
  # write each review to a separate row in the CSV file
  for review in reviews_df['Review Text']:
    # Apply preprocessing function to reviews
    temp = preprocess_text(review)
    count +=1
    writer.writerow({'review': temp})
    if count <= 15:
      print(str(count)+": "+temp+"\n")
print("\n")

## My competitor app2 Wechat
# Read in the CSV file containing the reviews
reviews_df = pd.read_csv('Task1/com.tencent.mm.csv', encoding='utf-8')

# open a CSV file and store pre-processed reviews 
with open('Task2/preprocessed_reviews_wechat.csv', mode='w', newline='', encoding='utf-8') as csv_file:
  fieldnames = ['review']
  writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
  writer.writeheader()
  # Count review number and for output 15 sample reviews
  count = 0
  print("My competitor app2 (Wechat) 15 sample pre-processed review:\n")
  # write each review to a separate row in the CSV file
  for review in reviews_df['Review Text']:
    # Apply preprocessing function to reviews
    temp = preprocess_text(review)
    count +=1
    writer.writerow({'review': temp})
    if count <= 15:
      print(str(count)+": "+temp+"\n")
```

```python
# Task 3
# Download libriry
!pip install textblob
!pip install vaderSentiment
!mkdir Task3
import textblob
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np

# Textblob analysis
def get_textblob_polarity(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    return polarity

# Vader analysis
def get_vader_polarity(text):
    analyzer = SentimentIntensityAnalyzer()
    scores = analyzer.polarity_scores(text)
    polarity = scores['compound']
    return polarity

# My SUD app (Birday) Sentiment Analysis
df_review = pd.read_csv('Task2/preprocessed_reviews_birday.csv')
# create two new DataFrame to store sentiment values
textblob_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'textblob_polarity'])
vader_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'vader_polarity'])
# iterate through each row in df and calculate the sentiment using TextBlob and Vader
for index, row in df_review.iterrows():
    # break if more than 1000 rows have been processed. Avoid run time too long
    if index >= 1000: 
        break
    app_name = "com.minar.birday"
    review = row['review']
    if isinstance(review, float) and np.isnan(review):
      # skip rows with NaN values in the 'Pre-processed reviews' column
      continue
    textblob_polarity = get_textblob_polarity(review)
    vader_polarity = get_vader_polarity(review)
    
    # append the calculated sentiment values to the new DataFrame
    textblob_sentiment_df = pd.concat([textblob_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'textblob_polarity': textblob_polarity}, index=[0])], ignore_index=True)
    vader_sentiment_df = pd.concat([vader_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'vader_polarity': vader_polarity}, index=[0])], ignore_index=True)

# store into CSV file
textblob_sentiment_df.to_csv('Task3/Birday_textblob_sentiment.csv', index=False)
vader_sentiment_df.to_csv('Task3/Birday_vader_sentiment.csv', index=False)
print("SUD app Birday sentiment analysis successed.")

# My similar app1 (KIC) Sentiment Analysis
df_review = pd.read_csv('Task2/preprocessed_reviews_kic.csv')
# create two new DataFrame to store sentiment values
textblob_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'textblob_polarity'])
vader_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'vader_polarity'])
# iterate through each row in df and calculate the sentiment using TextBlob and Vader
for index, row in df_review.iterrows():
    # break if more than 1000 rows have been processed. Avoid run time too long
    if index >= 1000: 
        break
    app_name = "com.kic.kicapp"
    review = row['review']
    if isinstance(review, float) and np.isnan(review):
      # skip rows with NaN values in the 'Pre-processed reviews' column
      continue
    textblob_polarity = get_textblob_polarity(review)
    vader_polarity = get_vader_polarity(review)
    
    # append the calculated sentiment values to the new DataFrame
    textblob_sentiment_df = pd.concat([textblob_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'textblob_polarity': textblob_polarity}, index=[0])], ignore_index=True)
    vader_sentiment_df = pd.concat([vader_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'vader_polarity': vader_polarity}, index=[0])], ignore_index=True)

# store into CSV file
textblob_sentiment_df.to_csv('Task3/kic_textblob_sentiment.csv', index=False)
vader_sentiment_df.to_csv('Task3/kic_vader_sentiment.csv', index=False)
print("Simillar app1 KIC sentiment analysis successed.")

# My similar app2 (youate) Sentiment Analysis
df_review = pd.read_csv('Task2/preprocessed_reviews_youate.csv')
# create two new DataFrame to store sentiment values
textblob_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'textblob_polarity'])
vader_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'vader_polarity'])
# iterate through each row in df and calculate the sentiment using TextBlob and Vader
for index, row in df_review.iterrows():
    # break if more than 1000 rows have been processed. Avoid run time too long
    if index >= 1000: 
        break
    app_name = "com.youate.android"
    review = row['review']
    if isinstance(review, float) and np.isnan(review):
      # skip rows with NaN values in the 'Pre-processed reviews' column
      continue
    textblob_polarity = get_textblob_polarity(review)
    vader_polarity = get_vader_polarity(review)
    
    # append the calculated sentiment values to the new DataFrame
    textblob_sentiment_df = pd.concat([textblob_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'textblob_polarity': textblob_polarity}, index=[0])], ignore_index=True)
    vader_sentiment_df = pd.concat([vader_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'vader_polarity': vader_polarity}, index=[0])], ignore_index=True)

# store into CSV file
textblob_sentiment_df.to_csv('Task3/youate_textblob_sentiment.csv', index=False)
vader_sentiment_df.to_csv('Task3/youate_vader_sentiment.csv', index=False)
print("Simillar app2 Youate sentiment analysis successed.")


# My competitor app1 (Discord) Sentiment Analysis
df_review = pd.read_csv('Task2/preprocessed_reviews_discord.csv')
# create two new DataFrame to store sentiment values
textblob_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'textblob_polarity'])
vader_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'vader_polarity'])
# iterate through each row in df and calculate the sentiment using TextBlob and Vader
for index, row in df_review.iterrows():
    # break if more than 1000 rows have been processed. Avoid run time too long
    if index >= 1000: 
        break
    app_name = "com.discord"
    review = row['review']
    if isinstance(review, float) and np.isnan(review):
      # skip rows with NaN values in the 'Pre-processed reviews' column
      continue
    textblob_polarity = get_textblob_polarity(review)
    vader_polarity = get_vader_polarity(review)
    
    # append the calculated sentiment values to the new DataFrame
    textblob_sentiment_df = pd.concat([textblob_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'textblob_polarity': textblob_polarity}, index=[0])], ignore_index=True)
    vader_sentiment_df = pd.concat([vader_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'vader_polarity': vader_polarity}, index=[0])], ignore_index=True)

# store into CSV file
textblob_sentiment_df.to_csv('Task3/discord_textblob_sentiment.csv', index=False)
vader_sentiment_df.to_csv('Task3/discord_vader_sentiment.csv', index=False)
print("Competitor app1 Discord sentiment analysis successed.")

# My competitor app2 (Wechat) Sentiment Analysis
df_review = pd.read_csv('Task2/preprocessed_reviews_wechat.csv')
# create two new DataFrame to store sentiment values
textblob_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'textblob_polarity'])
vader_sentiment_df = pd.DataFrame(columns=['app_package_name', 'review', 'vader_polarity'])
# iterate through each row in df and calculate the sentiment using TextBlob and Vader
for index, row in df_review.iterrows():
   # break if more than 1000 rows have been processed. Avoid run time too long
    if index >= 1000: 
        break
    app_name = "com.tencent.mm"
    review = row['review']
    if isinstance(review, float) and np.isnan(review):
      # skip rows with NaN values in the 'Pre-processed reviews' column
      continue
    textblob_polarity = get_textblob_polarity(review)
    vader_polarity = get_vader_polarity(review)
    
    # append the calculated sentiment values to the new DataFrame
    textblob_sentiment_df = pd.concat([textblob_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'textblob_polarity': textblob_polarity}, index=[0])], ignore_index=True)
    vader_sentiment_df = pd.concat([vader_sentiment_df, pd.DataFrame({'app_package_name': app_name, 'review': review, 'vader_polarity': vader_polarity}, index=[0])], ignore_index=True)

# store into CSV file
textblob_sentiment_df.to_csv('Task3/wechat_textblob_sentiment.csv', index=False)
vader_sentiment_df.to_csv('Task3/wechat_vader_sentiment.csv', index=False)
print("Competitor app2 Wechat sentiment analysis successed.")

print("All sentiment analysis output had successfully saved into csv file.")
```

```python
# Task 4
# For this task, I choose Discord.
#!pip install gensim
# load necessary libraries
import pandas as pd
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from gensim import corpora, models
from gensim.models.ldamodel import LdaModel
import os

# set the current working directory
os.chdir('/content')
# read in preprocessed reviews
df_review = pd.read_csv('Task2/preprocessed_reviews_discord.csv')

# skip NaN values in the 'review_text' column
documents = df_review['review'].dropna().tolist()

# create dictionary of words from the preprocessed reviews
dictionary = corpora.Dictionary([doc.split() for doc in documents])

# create document-term matrix
doc_term_matrix = [dictionary.doc2bow(doc.split()) for doc in documents]

# Build LDA model with 30 topics
lda_model = LdaModel(doc_term_matrix, num_topics=30, id2word=dictionary, passes=50, minimum_probability=0.01)

# print the topics and their descriptions
for idx, topic in lda_model.print_topics(num_topics=30, num_words=15):
    print(f'Topic {idx+1}: {topic}')
    words = [word.split('*')[1][1:-1] for word in topic.split('+') if len(word.split('*')[1][1:-1]) > 7]
    print(f'Description: {", ".join(words)}\n')
    
```
